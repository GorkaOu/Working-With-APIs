### https://walltime.info/api.html: we can get market info, in this case I get info of trades between BRL & XBT at 4 A.M with date 2018/07/30

```python
# Import requests library
import requests
import json
```

```python
response = requests.get('https://s3.amazonaws.com/data-production-walltime-info/production/dynamic/last-trades/v8878cb_BRL-XBT_2018_07_30_04_p0.json')
```

```python
print("Crypo:", response.status_code)
```

```python
#response
Crypto = response.json()
Crypto
```

```python
import pandas as pd
pd.set_option("display.max_rows", 0)

data = pd.DataFrame(Crypto)
data.head()
```

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        flattened.columns = [column + '_' + str(colname) for colname in flattened.columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
```

```python
nested_columns = ['last_trades']
```

```python
flatCrypto = flatten(data,nested_columns)
flatCrypto
```

### https://newsapi.org/v2/everything?domains=wsj.com&apiKey=f0891beff53347f297c4ce12326063d1: The newest articles The Wall Street Journal release

```python
url = 'https://newsapi.org/v2/everything?domains=wsj.com&apiKey=f0891beff53347f297c4ce12326063d1'

response = requests.get(url)

if response.status_code == 200:
    data2 = response.json()['articles']
    print('Successful request with status code',response.status_code)
else:
    print('Request failed with status code', response.status_code)
```
